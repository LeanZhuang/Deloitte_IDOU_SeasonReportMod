```python
import pandas as pd

# df1 = pd.read_excel('~/PycharmProjects/SeasonRepMod/整合季报代码模型/meta_code/后台_指标计算结果_20220926.xlsx')
df2 = pd.read_csv('~/PycharmProjects/SeasonRepMod/整合季报代码模型/后台指标原始数据/2022两个季度指标.csv')
df3 = pd.read_csv('~/PycharmProjects/SeasonRepMod/整合季报代码模型/后台指标原始数据/2022Q3季报指标.csv')
df_industry = pd.read_csv('/Users/zhuangyuhao/PycharmProjects/SeasonRepMod/整合季报代码模型/meta_code/fin_entity.csv')
```

```python
df20 = pd.read_csv('../后台指标原始数据/2020四个季度指标.csv')
df21 = pd.read_csv('../后台指标原始数据/2021四个季度指标.csv')
```

```python
df1 = pd.concat([df20, df21])
df1.drop('stock_name', inplace=True, axis=1)
df1.drop('factor_id', inplace=True, axis=1)
df1.drop('name', inplace=True, axis=1)
df1.rename(columns={'entity_name':'BASIC_entity_name',
                    'stock_code':'BASIC_code',
                    'year':'BASIC_year',
                    '季度':'BASIC_quarter',
                    'finnal_value':'end_value'
                    }, inplace=True)
df1.drop_duplicates(subset=["BASIC_entity_name", "BASIC_code", "BASIC_year", "BASIC_quarter", "code"], keep='first', inplace=True)
df1 = df1.pivot(index=["BASIC_entity_name", "BASIC_code", "BASIC_year", "BASIC_quarter"], #以主体划分
                columns="code",
                values="end_value").reset_index()
df1
```

```python
df1
```

```python
df_industry_keep = df_industry[['entity_name', 'industry_whitewash']]
df_industry_keep.drop_duplicates(inplace=True)
df_industry_keep.rename(columns={'entity_name':'BASIC_entity_name',
                                 'industry_whitewash':'BAISC_industry'},
                        inplace=True)
df_industry_keep
```

```python
df2.drop('stock_name', inplace=True, axis=1)
df2.drop('factor_id', inplace=True, axis=1)
df2.drop('name', inplace=True, axis=1)
```

```python
df2.rename(columns={'entity_name':'BASIC_entity_name',
                    'stock_code':'BASIC_code',
                    'year':'BASIC_year',
                    '季度':'BASIC_quarter',
                    'finnal_value':'end_value'
                    }, inplace=True)
df2.drop_duplicates(subset=["BASIC_entity_name", "BASIC_code", "BASIC_year", "BASIC_quarter", "code"], keep='first', inplace=True)
df2 = df2.pivot(index=["BASIC_entity_name", "BASIC_code", "BASIC_year", "BASIC_quarter"], #以主体划分
                columns="code",
                values="end_value").reset_index()
df2
```

```python
df3.drop('stock_name', inplace=True, axis=1)
df3.drop('factor_id', inplace=True, axis=1)
df3.drop('name', inplace=True, axis=1)
df3.rename(columns={'entity_name':'BASIC_entity_name',
                    'stock_code':'BASIC_code',
                    'year':'BASIC_year',
                    '季度':'BASIC_quarter',
                    'finnal_value':'end_value'
                    }, inplace=True)
df3.drop_duplicates(subset=["BASIC_entity_name", "BASIC_code", "BASIC_year", "BASIC_quarter", "code"], keep='first', inplace=True)
df3 = df3.pivot(index=["BASIC_entity_name", "BASIC_code", "BASIC_year", "BASIC_quarter"], #以主体划分
                columns="code",
                values="end_value").reset_index()
df3
```

```python
df_total = pd.concat([df1, df2, df3])
df_total.sort_values('BASIC_entity_name')
df_total.sort_values('BASIC_year')
df_total.sort_values('BASIC_quarter')
```

```python
df_total = df_total.drop(index=df_total[(df_total.BASIC_quarter == 0)].index.tolist())
df_total.reset_index()
# df_total.drop(columns='BASIC_industry', inplace=True)
```

```python
df_total_industry = df_total.merge(df_industry_keep, on='BASIC_entity_name', how='inner')
industry = df_total_industry.pop('BAISC_industry')
df_total_industry.insert(2, 'BASIC_industry', industry)
```
